```python
print('\n######### Correlation of each feature with the target class ##########\n')
corr_with_class = df.corr()['price'].drop('price')
print(corr_with_class)
```

```python
selected_features = df.corr()['price'].drop('price')[df.corr()['price'].abs() > 0.1].index.tolist()
print(selected_features)
```

```python
df = df[selected_features + ['price']]
df.to_csv('selected_features_on_kc_house_data.csv', index=False)
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
sm.graphics.plot_corr(df.corr(), xnames=list(df.corr()))
plt.show()
```

Dropping features just because they have low correlation (< 0.1) with the target can sometimes hurt the model, especially in nonlinear models.
It's important to know that I dropped because I want to test model with both type of dataset: - 1 with all features. - 1 with just those selected feature based on low correlation with target (if tree-based model doesn't help us to get a fair accuracy).
