```python
import pandas as pd
df = pd.read_csv('C:/Users/Admin/Downloads/kc_house_data.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=True, color='skyblue', bins=50)
```

```python
corr = df.corr(numeric_only=True)[['price']]
plt.figure(figsize=(12, 10))
sns.heatmap(corr.sort_values(by='price', ascending=False), annot=True, cmap='coolwarm')
```

```python
num_cols = ['sqft_living', 'sqft_above', 'sqft_living15', 'lat']

for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=col, y='price')
```

```python
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Price by Number of Bedrooms')
```

```python
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('Price by Number of Waterfront')
```

```python
sns.boxplot(x='floors', y='price', data=df)
plt.title('Price by Number of Floors')
```

```python
sns.boxplot(x='grade', y='price', data=df)
plt.title('Price by Number of Grades')
```

```python
sns.boxplot(x='view', y='price', data=df)
plt.title('Price by Number of View')
```

```python
sns.boxplot(x='condition', y='price', data=df)
plt.title('Price by Number of Condition')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='long', y='lat', hue='price', data=df, palette='viridis')
plt.title('House Prices by Location')
```

```python
cont_features = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'price', 'lat', 'long']

for col in cont_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pairplot_features = ['price', 'sqft_living', 'sqft_above', 'sqft_basement', 'lat', 'long', 'grade']

sns.pairplot(df[pairplot_features], diag_kind='kde', plot_kws={'alpha':0.5, 's':20})
plt.suptitle("Pairplot of Key Continuous Features with Price", y=1.02)
plt.show()
```
